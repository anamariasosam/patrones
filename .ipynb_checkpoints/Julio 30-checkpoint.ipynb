## Modelo biclase
Me sireve para saber si un modelo está bien clasificado

- Sensibilidad: es buena cuando es cercana a 1 o 1  . No comete errores clasificando los positivos
- Especificidad: que tambien clasifico los negativos

- Precision: 

FN: Le dijo que no tenía cancer y si tenía

# Metodologías de validación
Una estrategia para particionar la base de datos, para tener una parte con la que entrenemos el modelo y otra con la cual validar.
    
## Generalización

Esta capacidad de clasificar correctamente muestras nuevas.

Garantizando la generelización

## Validacion cruzada, cross validation

Divide en manera aleatoria
k el numero de grupos que saca, también folds
k-1 entrena el modelo

Las iteraciones son iguales a los folds

Itero y saco promedio de los especificidad,eficiencia.

## Validacion cruzada, leave one out

Entreno con n - 1 y valido con 1


## Validacion Bootstrap
Es con porcentaje 70%  30%

El numero promedio de iteraciones para unas 300 muestras la iteracion debe ser de 10 repeticiones
